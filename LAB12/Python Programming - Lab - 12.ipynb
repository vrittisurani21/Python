### 04) WAP to demonstrate the use of Pie chart.

```python
Subjects = ["Physics", "Chemistry", "Biology", "English", "Sanskrit"]
Marks = [30, 28, 20, 25, 28]  
plt.pie(Marks, labels=Subjects,autopct="%1.1f%%") 
plt.axis("equal") 
plt.title("subject Result Distribution")
plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python

Subjects = ['DBMS', 'DS', 'PS', 'DM','DF','BM']
Marks = [90, 80, 50, 100,40,70]
plt.bar(Subjects, Marks,color='orange')
plt.title('Result')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.show()
```

### 07) WAP to demonstrate the use of Histogram.

```python
import random
import matplotlib.pyplot as plt

# Generate random data
data = [random.random() for _ in range(1000)]

# Plot histogram
plt.hist(data, bins=30, color='cyan', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()

```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
SUBJECT = ['BIOLOGY', 'CHEMISTRY', 'PHYSICS', 'MATHS']
MARKS = [45, 40, 50, 30]
bars = plt.bar(SUBJECT, MARKS,color='grey')
plt.bar_label(bars)
plt.xlabel("SUBJECT")
plt.ylabel("MARKS")
plt.title("Bar Chart with Value Labels")
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import random
x = [random.random() for _ in range(50)]
y = [random.random() for _ in range(50)]
colors = [random.random() for _ in range(50)]
plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```
